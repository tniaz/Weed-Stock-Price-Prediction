```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

```python
#Reading csv files into dataframe
sns.set(style='whitegrid',palette='Set2')
a=pd.read_csv('weedstockdata.csv')
b=pd.read_csv('newsscores.csv')


```

```python
#Plotting data
a.plot(x='Date',y='Adj Close',) 
plt.title('Price vs. Date', color='black')
a.plot(x='Date',y='Volume',) 
plt.title('Volume vs. Date', color='black')
b.plot(x='date',y='polarity',) 
plt.title('Polarity vs. Date', color='black')
b.plot(x='date',y='subjectivity',) 
plt.title('Subjectivity vs. Date', color='black')
plt._show()

##Notes: To add date axis limit and formatting from csv data.
```
