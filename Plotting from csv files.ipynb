Importing required packages

```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
```

Plotting

```python
#function to plot data from the two csv files
def plot (weedstockcsv, newsscorecsv):
    a=pd.read_csv('weedstockdata.csv')
    b=pd.read_csv('newsscores.csv')
    sns.set(style='whitegrid',palette='Set2')
    x=a.index
    labels = a['Date']
    a.plot(x='Date',y='Adj Close',figsize=(14,7)) 
    plt.title('Price vs. Date', color='black',fontsize=14)
    plt.xticks(x, labels, rotation = 'vertical')
    plt.ylabel('BTC Price(USD)', fontsize=10)
    a.plot(x='Date',y='Volume',figsize=(14,7)) 
    plt.title('Volume vs. Date', color='black',fontsize=14)
    plt.xticks(x, labels, rotation = 'vertical')
    plt.ylabel('Volume', fontsize=10)
    x=b.index
    labels = b['date']
    b.plot(x='date',y='polarity',figsize=(14,7)) 
    plt.title('Polarity vs. Date', color='black',fontsize=14)
    plt.xticks(x, labels, rotation = 'vertical')
    plt.ylabel('Polarity', fontsize=10)
    b.plot(x='date',y='subjectivity',figsize=(14,7))
    plt.title('Subjectivity vs. Date', color='black',fontsize=14)
    plt.xticks(x, labels, rotation = 'vertical')
    plt.ylabel('Subjectivity', fontsize=10)
    plt._show()
```

```python
plot ('weedstockdata.csv', 'newsscores.csv')
```
